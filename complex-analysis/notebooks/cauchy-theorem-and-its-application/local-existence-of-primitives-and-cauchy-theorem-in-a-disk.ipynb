# Local Existence of Primitives and Cauchy's Theorem in a Disk

```julia
using Integrals
using Statistics
using Plots
using LaTeXStrings

include("../curve_plot.jl")
include("../line_segment_plot.jl")

const FIGURES_DIR = "../../figures/"
```

```julia
z = 1 + 3im
h = 1.2 + 1.6im

p1 = plot(
    aspect_ratio=:equal,
    border=:none,
    legend=nothing,
)

plot_line_segment!(
    0.0 + 0.0im,
    real(z + h) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z + h) + 0.0im,
    z + h,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z) + 0.0im,
    0.0 + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
    color=:blue,
    linestyle=:dash,
)

plot_line_segment!(
    z,
    real(z) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
    color=:blue,
    linestyle=:dash,
)

annotate!(
    0 - 0.2,
    0,
    text(L"0"),
)

annotate!(
    real(z) - 0.2,
    imag(z) - 0.2,
    text(L"z"),
)

annotate!(
    real(z + h) + 0.2,
    imag(z + h) + 0.2,
    text(L"z + h"),
)

p2 = plot(
    aspect_ratio=:equal,
    border=:none,
    legend=nothing,
)

plot_line_segment!(
    z,
    real(z) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z) + 0.0im,
    real(z + h) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z + h) + 0.0im,
    z + h,
    which_arrow=:right,
    arrow_distance=0.2,
)

annotate!(
    real(z) - 0.2,
    imag(z) - 0.2,
    text(L"z"),
)

annotate!(
    real(z + h) + 0.2,
    imag(z + h) + 0.2,
    text(L"z + h"),
)

p3 = plot(
    aspect_ratio=:equal,
    border=:none,
    legend=nothing,
)

plot_line_segment!(
    z,
    real(z) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z) + 0.0im,
    real(z + h) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z + h) + 0.0im,
    z + h,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    z,
    real(z + h) + im * imag(z),
    which_arrow=:right,
    arrow_distance=0.2,
    color=:blue,
    linestyle=:dash,
    arrow_length=0.4,
)

plot_line_segment!(
    real(z + h) + im * imag(z),
    real(z + h) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
    color=:blue,
    linestyle=:dash,
)

plot_line_segment!(
    real(z + h) + 0.0im,
    real(z) + 0.0im,
    which_arrow=:right,
    arrow_distance=0.2,
    color=:blue,
    linestyle=:dash,
    arrow_length=0.4,
)

plot_line_segment!(
    real(z) + 0.0im,
    z,
    which_arrow=:right,
    arrow_distance=0.2,
    color=:blue,
    linestyle=:dash,
)

annotate!(
    real(z) - 0.2,
    imag(z) - 0.2,
    text(L"z"),
)

annotate!(
    real(z + h) + 0.2,
    imag(z + h) + 0.2,
    text(L"z + h"),
)

p4 = plot(
    aspect_ratio=:equal,
    border=:none,
    legend=nothing,
    ylim=(1, 5),
)

plot_line_segment!(
    z,
    real(z + h) + im * imag(z),
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z + h) + im * imag(z),
    z + h,
    which_arrow=:right,
    arrow_distance=0.2,
)

annotate!(
    real(z) - 0.2,
    imag(z) - 0.2,
    text(L"z"),
)

annotate!(
    real(z + h) + 0.2,
    imag(z + h) + 0.2,
    text(L"z + h"),
)

p5 = plot(
    aspect_ratio=:equal,
    border=:none,
    legend=nothing,
    ylim=(1, 5),
)

plot_line_segment!(
    z,
    real(z + h) + im * imag(z),
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    real(z + h) + im * imag(z),
    z + h,
    which_arrow=:right,
    arrow_distance=0.2,
)

plot_line_segment!(
    z,
    z + h,
    which_arrow=:right,
    arrow_distance=0.1,
    color=:blue,
    linestyle=:dash,
    arrow_length=0.3,
)

plot_line_segment!(
    z + h,
    real(z + h) + im * imag(z),
    which_arrow=:right,
    arrow_distance=0.1,
    color=:blue,
    linestyle=:dash,
    arrow_length=0.3,
)

plot_line_segment!(
    real(z + h) + im * imag(z),
    z,
    which_arrow=:right,
    arrow_distance=0.1,
    color=:blue,
    linestyle=:dash,
    arrow_length=0.3,
)

annotate!(
    real(z) - 0.2,
    imag(z) - 0.2,
    text(L"z"),
)

annotate!(
    real(z + h) + 0.2,
    imag(z + h) + 0.2,
    text(L"z + h"),
)

p6 = plot(
    aspect_ratio=:equal,
    border=:none,
    legend=nothing,
    ylim=(1, 5),
)

plot_line_segment!(
    z,
    z + h,
    which_arrow=:left,
    arrow_distance=0.2,
)

annotate!(
    real(z) - 0.2,
    imag(z) - 0.2,
    text(L"z"),
)

annotate!(
    real(z + h) + 0.2,
    imag(z + h) + 0.2,
    text(L"z + h"),
)

annotate!(
    real(z + h / 2) + 0.2,
    imag(z + h / 2) - 0.2,
    text(L"\eta"),
)

plot(
    p1, p2, p3, p4, p5, p6,
    layout=(2, 3),
    link=:all,
)
```
