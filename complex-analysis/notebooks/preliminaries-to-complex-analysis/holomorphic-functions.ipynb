# Holomorphic Functions

```julia
using Plots
using LaTeXStrings

const FIGURES_DIR = "../../figures/"
```

```julia
x = -1:0.001:1
f = x -> cbrt(x)

plot(
    aspect_ratio=:equal,
)

plot!(
    x,
    f.(x),
    label="\$f(x) = \\sqrt[3]{x}\$",
)

vline!(
    [0],
    label="vertical tangent line at \$x=0\$",
)

```

```julia
savefig(joinpath(FIGURES_DIR, "cubic-root-function.svg"))
```
