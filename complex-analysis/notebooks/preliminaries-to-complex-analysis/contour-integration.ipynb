# Contour Integration

```julia
using Integrals
using Plots
using LaTeXStrings

const FIGURES_DIR = "../../figures/"
```

```julia
f = x -> x > 0 ? 2x * sin(1 / x) - cos(1 / x) : 0.0
problem = IntegralProblem((x, p) -> f(x), 0.0, 2 / pi)
solution = solve(problem, QuadGKJL())
```

```julia
F = x -> x > 0 ? x^2 * sin(1 / x) : 0.0
F(2 / pi) - F(0)
```

```julia
solution.u ≈ F(2 / pi) - F(0) ≈ 4 / pi^2
```

```julia
x = 0:0.00001:2/pi

p1 = plot(
    x,
    f.(x),
    color=palette(:default)[1],
)

p2 = plot(
    x,
    F.(x),
    color=palette(:default)[2],
)

plot(
    p1,
    p2,
    layout=(1, 2),
    legend=nothing,
)
```
