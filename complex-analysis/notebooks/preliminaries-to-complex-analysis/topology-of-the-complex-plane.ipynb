# Topology of the Complex Plane

```julia
using Plots
using LaTeXStrings
include("../curve_plot.jl")

const FIGURES_DIR = "../../figures/"
```

```julia
alpha = t -> t^3 - t^2 + im * t^3
a = -0.7
b = 1.0
z0 = alpha(a)
z = alpha(b)

plot(
    aspect_ratio=:equal,
    border=:none,
)

plot_curve!(
    alpha,
    a,
    b,
    arrow_points=[-0.5, 0.7]
)


 = 0:0.01:2pi
plot!(
    real(z .+ 0.5 * exp.(im * )),
    imag(z .+ 0.5 * exp.(im * )),
    color=:black,
    linestyle=:dash,
    label=nothing
)

w = z + 0.3 * exp(im * 3pi / 4)
beta = t -> (1 - t) * z + t * w
plot_curve!(
    beta,
    0.0,
    1.0,
    arrow_points=[0.5]
)

annotate!(
    real(z0) - 0.1,
    imag(z0) - 0.1,
    text(L"z_0")
)

annotate!(
    real(z) + 0.1,
    imag(z),
    text(L"z")
)

annotate!(
    real(w) - 0.05,
    imag(w) + 0.05,
    text(L"w")
)

annotate!(
    real(z) - 0.7,
    imag(z) - 0.3,
    text(L"D(z, r)")
)

annotate!(
    real(alpha(0)) + 0.1,
    imag(alpha(0)) - 0.1,
    text(L"\alpha")
)

annotate!(
    real(beta(0.5)) - 0.1,
    imag(beta(0.5)) - 0.1,
    text(L"\beta")
)
```

```julia
savefig(joinpath(FIGURES_DIR, "piecewise-smooth-curve-from-z0-to-w.svg"))
```
